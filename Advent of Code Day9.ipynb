```python
file = open("AoCDay9.txt","r")

map = file.read().split("\n")

shape = [len(map),len(map[0])]

lowpoint = ["lower","lower","lower","lower"]
lowpoints = []

for x in range(len(map)):
    for y in range(len(map[x])):
        coords = int(map[x][y])
        point = ["","","",""]
        if x == 0:
            point[0] = "lower"
        else:
            up = int(map[x-1][y])
            if coords < up:
                point[0] = "lower"
        
        if x == len(map)-1:
            point[1] = "lower"
        else:
            down = int(map[x+1][y])
            if coords < down:
                point[1] = "lower"
        
        if y == 0:
            point[2] = "lower"
        else:
            left = int(map[x][y-1])
            if coords < left:
                point[2] = "lower"
        
        if y == len(map[0])-1:
            point[3] = "lower"
        else:
            right = int(map[x][y+1])
            if coords < right:
                point[3] = "lower"
            
        if point == lowpoint: 
            lowpoints.append(coords)   

print(sum(lowpoints) + len(lowpoints))
```
